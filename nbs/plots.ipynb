```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Adapa_FH_task5_1': [0.718, 0.631, 0.860],
    'Adapa_FH_task5_2': [0.723, 0.745, 0.847],
    'Baseline': [0.619, 0.664, 0.742]
}, index=['Macro-AUPRC', 'Micro-F1', 'Micro-AUPRC'])
```

```python
df
```

```python
from plotnine import *
%matplotlib inline
```

```python
toplot = df.reset_index().melt('index').loc[lambda x: x.variable.isin(['Baseline', 'Adapa_FH_task5_1'])]
toplot['label_pos'] = toplot.value + 0.05
```

```python
(ggplot(toplot)
 + geom_bar(aes('index', 'value', fill='variable'), position='dodge', stat='identity')
 + coord_flip()
 + theme_bw()
 + theme(legend_position='top')
 + scale_fill_discrete(name=' ')
 + xlab('') + ylab(' ')
 + scale_y_continuous(expand=(0, 0), limits=(0, 1))
 + geom_text(aes('index', 'label_pos', label='value', group='variable'), position=position_dodge(width=1))
)
```
